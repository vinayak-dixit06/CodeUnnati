### Practical 2 : Syncing NumPy and Matplotlib and printing a line

```python
# Aim : Plotting a line using NumPy and Matplotlib

# Importing numpy library
import numpy as np
# Importing matplotlib library and pyplot sub-library
import matplotlib.pyplot as plt

# Taking data co-ordinates
ypoints = np.array([3,8,1,10])
# As we have defined the y co-ordinates in terms of an array, the x co-ordinates by-default will be the indices of the array.
# Therefore, xpoints = [0,1,2,3]

# Plotting the co-ordinates
# marker attribute defines the data points which can be "o","*","."
# color attribute defines the color of the line which can be any of the valid colors
plt.plot(ypoints, marker="o", color="indigo")

# Printing the output
plt.show()
```

### Practical 3 : Labelling the axes and chart title

```python
# Aim : Labelling the chart and both the axis

# Importing numpy library
import numpy as np
# Importing matplotlib library and pyplot sub-library
import matplotlib.pyplot as plt

# Taking x co-ordinate values
xpoints = np.array([2,3,8,7])
# Taking y co-ordinate values
ypoints = np.array([1,4,1,9])

# Plotting the line
# linestyle attribute changes the style of the plotted line to any valid style which can be "-.","--","dotted"
plt.plot(xpoints,ypoints, marker="*", color="blue", linestyle = "dotted")

# for labelling X-axis, use the xlabel() method
plt.xlabel("X-axis")
# for labelling Y-axis, use the ylabel() method
plt.ylabel("Y-axis")
# for labelling the title of the chart, use the title() method
plt.title("GRAPH")

# Printing the output
plt.show()
```

### Practical 4 : Plot a line chart describing marks of students of a given class

```python
# Aim : Plotting a line chart describing marks of student of a given class

# Importing matplotlib library and pyplot sub-library
import matplotlib.pyplot as plt
# Importing numpy library and random sub-library
from numpy import random as random

# Creating list of Students 
students = ["Shubham","Raj","Suketu","Nishang","Vatsal","Jay","Vinayak","Jyot"]

# Defining marks of Students
marks=[]
for i in range(0,len(students)) :
    marks.append(random.randint(0,101))

# Labelling X-axis
plt.xlabel("Students")
# for giving the inclined effect to the labels of x-axis, use the method the xticks(rotation = float) method
plt.xticks(rotation=45)
# Labelling Y-axis
plt.ylabel("Marks")
# Labelling the Chart
plt.title("SEM 7")

# Plotting the co-ordinates
# markerfacecolor : markerfacecolor attributes changes the color of the data-points to any valid color
# markersize : markersize attribute changes the size of all data-points to as per the defined size
plt.plot(students, marks, marker="o", color="black", markerfacecolor="blue", markersize=10)

# Printing the output
plt.show()
```

## 3.4. Bar Chart :

    A bar graph is a chart that represents categorical data with rectangular bars with height or length proportional to the
    values that they represent
    
    Syntax :
        ax.bar(x, height, width, bottom, align)
        plt.bar(x,y)
        
    Bar graphs are of two types :
        1. Vertical
        2. Horizontal
        
    By default the type of bar chart is vertical

### Practical 9 : Plotting a Bar chart showing the Quaterly Revenue and Quaterly Profit of TECHM Ltd. for the financial year 2022.

```python
# Aim : Plot a Bar chart showing the Quaterly Revenue and Quaterly Profit of Tech Mahindra Ltd. for the financial year 2022

# Importing matplotlib library and pyplot sub-library
import matplotlib.pyplot as plt

# Taking x co-ordinates
quaters = ["June-21","September-21","December-21","March-22"]
# Taking y co-ordinates
revenue = [10198,10881,11451,12116]
profit = [1353,1339,1369,1503]

fig1 = plt.figure(figsize = (8,6))
# Labelling X-axis
plt.xlabel("Quaters")
# Labelling Y-axis
plt.ylabel("Rs (in Cr)")
# Labelling Title of the Chart
plt.title("TECHM FY - 22")

# Plotting the co-ordinates

# for adding value labels
for i in range(len(quaters)) :
    plt.text(quaters[i],revenue[i],revenue[i],ha="center",va="bottom", fontweight="bold", fontsize = 12)
# Plotting the revenue
plt.bar(quaters, height= revenue, width = 0.4, edgecolor="green", color="lightgreen")

# for adding value labels
for i in range(len(quaters)) :
    plt.text(quaters[i], profit[i], profit[i], ha="center", va="bottom", fontweight="bold", fontsize = 12)
# Plotting the Profit
plt.bar(quaters, height =profit, width = 0.4, edgecolor="blue", color="lightblue")

# Adding legend for better data interpretation
plt.legend(["Revenue","Profit"], loc = "upper left")

# Printing the output
plt.show()
```

### Practical 11 : Manipluating tick parameters of the axes

```python
# Ticks are the markers denoting data points on either axis
# Importing matplotlib library and pyplot sub-library
import matplotlib.pyplot as plt

# Taking Co-ordinates of X and Y
x = [5,10,15,20,25,30,35,40,45,50]
y = [1,4,3,2,7,6,9,8,10,5]

# Plotting the co-ordinates
plt.plot(x,y,"g")

# Labelling x-axis
plt.xlabel("X-axis")
# Labelling y-axis
plt.ylabel("Y-axis")

# Manipulating tick parameters

# labelsize : labelsize attribute is used to change the size of the labels
# labelrotation : labelrotation attribute is used to give a definite inclined angle to the labels
# labelcolor : labelcolor attribute is used to change the color of the label

plt.tick_params(axis="x", labelsize = 12, labelrotation = 20, labelcolor = "red")
plt.tick_params(axis="y", labelsize = 12, labelrotation = 20, labelcolor = "blue")

# Printing the output
plt.show()
```

### Practical 14 : Annotations and Labeling

```python
# Aim : Adding Annotations to a chart

# Importing matplotlib library and pyplot sub-library
import matplotlib.pyplot as plt
# Importing numpy library
import numpy as np

# Draw plot as figure
fig, ax = plt.subplots()

# Taking x co-ordinates
x = np.linspace(0,20,1000)
# Plotting the co-ordinates
ax.plot(x,np.cos(x))
ax.axis("equal")

# Annoting the local maximum point
ax.annotate('local maximum', xy = (6.8,1), xytext = (10,4), arrowprops = dict(facecolor="black", shrink=0.05))

# Annoting the local minimum point
ax.annotate('local minimum', xy = (5* np.pi,-1), xytext=(2,-6),arrowprops=dict(arrowstyle='->', connectionstyle="angle3,angleA=0,angleB=-90"))

# Printing the output
plt.show()
```

### Practical 15 : Adding Annotations and Legend

```python
# Aim : Adding Annotation and Legend

# Importing numpy library
import numpy as np
# Importing matplotlib library and pyplot sub-library
import matplotlib.pyplot as plt

# You can also write
# from matplotlib import pyplot as plt

# Taking x co-ordinates
x = np.arange(0,10)
# Taking y co-ordinates
y = x ^ 2
# Taking z co-ordinates
z = x ^ 3

# Plotting the co-ordinates of x and y
plt.plot(x,y,"red", marker="o")
# Plotting the co-ordinates of x and z
plt.plot(x,z,"green")
# Plotting the co-ordinates of y and z
plt.plot(y,z,"blue")

# Adding Title
plt.title("Graph")
# Adding X-axis label
plt.xlabel("Time")
# Adding Y-axis label
plt.ylabel("Distance")

# Adding Legend
plt.legend(["Race 1","Race 2","Race 3"],loc="lower right")

# Annotations
plt.annotate(xy=[0.2,0.2], text = "First Entry", rotation = "20")
plt.annotate(xy=[2.5,6.5], text = "Second Entry", rotation = "20")
plt.annotate(xy=[6.3,10.5], text = "Third Entry", rotation = "20")

# Printing the output
plt.show()
```

## 3.7. Pie Plot :

    Pie chart is a circular statistical plot that can dispplay only one series of data

    The area of the chart is total % of the given data
    
    Syntax :
        plt.pie(data)

-------------------------------------------------------------------------------------------------------------------------------
### Practical 17 : Visualizing Portfolio allocation report using Pie plot

```python
# Aim : Create a stocks portfoilo using pie plot

# Importing NumPy library
import numpy as np
# Importing Matplotlib library and pyplot sub-library
import matplotlib.pyplot as plt

# Plot 1
# Taking the data co-ordinates (slices of the pie chart)
Stocks = ["WIPRO","HDFCBANK","BPCL","SAIL","PEL","TANLA","TATAMTRDVR","INFY","BANCOINDIA","TATAMOTORS","SOUTHBANK"]
# Taking the actual data-points
data1 = [11674.08,7515,6985.8,6900,5750,5226,4964,4830,2349,1420,1030]
# To highlight a slice the explode attribute is used
explode1 = (0.05,0.01,0.01,0,0.01,0.01,0.01,0.01,0.01,0.01,0.01)
# Adjusting figure size using figure(width, height) method
fig1 = plt.figure(figsize=(15,15))
# Subplotting the figure in Plot 1 using subplot(no. of rows, no. of cols, plot no.) method
plt.subplot(1,2,1)
# Labelling the chart
title1 = plt.title("STOCK-WISE ALLOCATION")
# Plotting the Pie chart
plt.pie(data1,labels=Stocks, explode = explode1, autopct="%1.1f%%")


# Plot 2
# Taking the data co-ordinates (slices of the pie chart)
sectors = ["INFORMATION & TECHNOLOGY","BANKS","OIL & GAS","METALS","AUTOMOBILES","FINANCIAL SERVICES","ENGINEERING PRODUCTS"]
# Taking the actual data-points
data2 = [21730.08,8545,6985.8,6900,6384,5750,2349]
# To highlight a slice the explode attribute is used
explode2 = (0.05,0.01,0.01,0.01,0.01,0.01,0.01)
# Adjusting figure size using figure(width, height) method
fig2 = plt.figure(figsize=(15,15))
# Subplotting the figure in Plot 2 using subplot(no. of rows, no. of cols, plot no.) method
plt.subplot(1,2,2)
# Labelling the chart
title2 = plt.title("SECTOR-WISE ALLOCATION")
# Plotting the Pie chart
plt.pie(data2, labels = sectors, autopct = "%1.1f%%", explode = explode2)

# Printing the output
plt.show()
```

## 3.8. 3-Dimensional Plotting :

    For visualling the data which has three dimensions, we require 3-Dimensional plotting
    
    In order to plot 3D data, we need to import mpl_toolkits library and mplot3d sub library
    
    Syntax :
        import mpl_toolkits.mplot3d  OR  from mpl_toolkits import mplot3d
    
    For Plotting the values use the plot3D method
    
    Syntax :
        ax = plt.axes(projection = "3d")
        ax.plot3D(x,y,z)

```python
# Aim : 3-Dimensional Plotting

# Importing numpy library
import numpy as np
# Importing matplotlib library and pyplot sub-library
import matplotlib.pyplot as plt
# Importing mpl_toolkits library and mplot3d sub-library
from mpl_toolkits import mplot3d

# Adjusting the figure
fig = plt.figure(figsize = (8,6))

# Syntax :
ax = plt.axes(projection="3d")

# Defining z-axis
z = np.linspace(0,1,100)
# Defining x-axis
x = z * np.sin(25 * z)
# Defining y-axis
y = z * np.cos(25 * z)


# Plotting the co-ordinates
ax.plot3D(x,y,z,"g")

# Labelling the Chart
ax.set_title("3D Plotting")

# To print the chart use the command
# plt.savefig("class.pdf",format="pdf")

# Printing the Output
plt.show()
```

### Methods and thier attribures :

    A). plot() : plot() method is used to print a line chart with the given co-ordinates
        Attributes : 
            1. linestyle : linestyle attribute is used to change the style of line
                example : linestyle = "dotted"
            2. marker : marker attribute is used to change the style of data-points
                example : marker = "o"
            3. markerfacecolor = markerfacecolor attribute is used to change the color of the marker
                example : markerfacecolor = "blue"
            4. color : color attribute is used to change the color of the plotted line
                example : color = "blue"
            5. markersize : markersize attribute is used to change the size of the marker
                 example : markersize = 12

    B). bar() : bar() method is used to print a bar chart with the given categorical data
                bar() : bar() method is used to print a vertical bar graph
                barh() : barh() method is used to print a horizontal graph
        Attributes :
            1. edgecolor : edgecolor attribute is used to change the color of the edge
                example : edgecolor = "blue"
            2. color : color attribute is used to change the color of the bar
                example : color = "lightblue"
            3. width : width attribute is used to change the width of bar
                example : width = 0.4
            4. height : height attribute is used to set the height of the bar
                example : height = y
            5. fontweight : fontweight is used to change the weight of the label text
                example : fontweight = "bold"
            6. fontsize : fontsize attribute is used to change the size of the label text
                example : fontsize = 12
            7. horizontalalign = horizontal align attribute is used to set the horizontal alignment of the label text
                example : horizontalalign = "center"
            8. verticalalign : vertical align attribute is used to set the vertical alignment of the label text
                example : verticalalign = "bottom"
